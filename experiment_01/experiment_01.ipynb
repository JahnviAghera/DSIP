#### Unit Impulse Signal

```python
import numpy as np
import matplotlib.pyplot as plt

def unit_impulse(length, position):
    signal = np.zeros(length)
    signal[position] = 1
    return signal

# Parameters
start = -10  # Start value of the x-axis range
stop = 10  # Stop value of the x-axis range
step = 1  # Step size

# Generate x-axis values
x = np.arange(start, stop+step, step)

# Generate unit impulse signal
impulse_signal = unit_impulse(len(x), abs(start)//step)

# Plot the signal
plt.stem(x, impulse_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Unit Impulse Signal')
plt.grid(True)
plt.show()
```

### AIM : Stimulate Impulse Train

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_impulse_train(signal_length, period):
    impulse_train = np.zeros(signal_length)
    for n in range(signal_length):
        if n % period == 0:
            impulse_train[n] = 1
    return impulse_train

# Define the parameters for the impulse train
signal_length = 100  # Length of the impulse train
period = 10  # Period of the impulse train

# Simulate the impulse train
impulse_train = simulate_impulse_train(signal_length, period)

# Plot and display the impulse train
plt.stem(impulse_train)
plt.title('Impulse Train')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.show()

# Save the impulse train array (optional)
# np.savetxt('impulse_train.txt', impulse_train, delimiter=',')

```

### **AIM: Write a python program to Simulate continuous and discrete unit step signal**

### Continuous and Discrete Unit Step Signals

The **unit step signal** is a fundamental signal in signal processing and system analysis. It is widely used to represent systems that switch on at a certain point in time.

#### Continuous Unit Step Signal

The continuous unit step signal, denoted as \( u(t) \), is defined as:
$$
u(t) = 
\begin{cases}
0, & t < 0 \\
1, & t \geq 0
\end{cases}
$$

- **Behavior:** The signal remains at zero for all negative time values and jumps to one at \( t = 0 \), remaining at one for all positive time values.
- **Applications:** Used to model systems that start operating at \( t = 0 \), such as switching circuits and control systems.

#### Discrete Unit Step Signal

The discrete unit step signal, denoted as \( u[n] \), is defined as:
$$
u[n] = 
\begin{cases}
0, & n < n_0 \\
1, & n \geq n_0
\end{cases}
$$

- **Behavior:** The signal is zero for all indices less than \( n_0 \) (often the midpoint of the sample range) and one for \( n \geq n_0 \).
- **Applications:** Used in digital signal processing to represent the activation of a system at a specific sample.

#### Visualization

- The **continuous unit step** appears as a horizontal line at zero for \( t < 0 \) and jumps to one at \( t = 0 \).
- The **discrete unit step** is shown as a sequence of zeros followed by ones, with the transition at the midpoint sample.

#### Importance

Unit step signals are essential for analyzing system responses, especially in determining how systems react to sudden changes or inputs. They are also used in constructing more complex signals and in solving differential and difference equations.

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_continuous_unit_step(time):
    unit_step = np.zeros_like(time)
    unit_step[time >= 0] = 1
    return unit_step

def simulate_discrete_unit_step(num_samples):
    unit_step = np.zeros(num_samples)
    unit_step[num_samples // 2:] = 1
    return unit_step

# Define the time range for the continuous unit step signal
time = np.linspace(-5, 5, 1000)  # Time range from -5 to 5

# Simulate the continuous unit step signal
continuous_unit_step = simulate_continuous_unit_step(time)

# Define the number of samples for the discrete unit step signal
num_samples = 20  # Number of samples

# Simulate the discrete unit step signal
discrete_unit_step = simulate_discrete_unit_step(num_samples)

# Plot and display the continuous and discrete unit step signals
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(time, continuous_unit_step)
plt.title('Continuous Unit Step Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.stem(discrete_unit_step)
plt.title('Discrete Unit Step Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

# Save the unit step signal arrays (optional)
# np.savetxt('continuous_unit_step.txt', continuous_unit_step, delimiter=',')
# np.savetxt('discrete_unit_step.txt', discrete_unit_step, delimiter=',')
```

### **AIM: Write a python program to Simulate continuous and discrete ramp signal**

### Explanation: Continuous and Discrete Ramp Signals

**Ramp signals** are fundamental in signal processing and system analysis, representing a signal that increases linearly with time.

#### Continuous Ramp Signal

A continuous ramp signal is mathematically represented as:

$$
x(t) = 
\begin{cases}
0, & t < 0 \\
a t, & t \geq 0
\end{cases}
$$

where:
- \( a \) is the slope of the ramp,
- \( t \) is the continuous time variable.

- For \( t < 0 \), the signal is zero.
- For \( t \geq 0 \), the signal increases linearly with time.

#### Discrete Ramp Signal

A discrete ramp signal is given by:

$$
x[n] = 
\begin{cases}
0, & n < n_0 \\
a (n - n_0), & n \geq n_0
\end{cases}
$$

where:
- \( n \) is the discrete time index,
- \( n_0 \) is the starting index (often the midpoint of the sample range),
- \( a \) is the slope.

- For \( n < n_0 \), the signal is zero.
- For \( n \geq n_0 \), the signal increases linearly with the sample index.

#### Applications

- Ramp signals are used to model systems with constant acceleration, such as velocity in uniformly accelerated motion.
- They are also used in control systems, testing system responses, and digital signal processing.

#### Visualization

- The **continuous ramp signal** appears as a straight line starting from zero at \( t = 0 \) and increasing with time.
- The **discrete ramp signal** is shown as a sequence of points (stems) that start from zero and increase linearly from the midpoint sample.

Ramp signals are essential for understanding system dynamics and are widely used in both theoretical and practical engineering applications.

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_continuous_ramp(time, slope):
    ramp = np.zeros_like(time)
    ramp[time >= 0] = slope * time[time >= 0]
    return ramp

def simulate_discrete_ramp(num_samples, slope):
    ramp = np.zeros(num_samples)
    ramp[num_samples // 2:] = slope * np.arange(num_samples // 2, num_samples)
    return ramp

# Define the time range for the continuous ramp signal
time = np.linspace(-5, 5, 1000)  # Time range from -5 to 5

# Define the number of samples and slope for the discrete ramp signal
num_samples = 20  # Number of samples
slope = 2  # Slope of the ramp

# Simulate the continuous ramp signal
continuous_ramp = simulate_continuous_ramp(time, slope)

# Simulate the discrete ramp signal
discrete_ramp = simulate_discrete_ramp(num_samples, slope)

# Plot and display the continuous and discrete ramp signals
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(time, continuous_ramp)
plt.title('Continuous Ramp Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.stem(discrete_ramp)
plt.title('Discrete Ramp Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

# Save the ramp signal arrays (optional)
# np.savetxt('continuous_ramp.txt', continuous_ramp, delimiter=',')
# np.savetxt('discrete_ramp.txt', discrete_ramp, delimiter=',')
```

### **AIM: Write a python program to Simulate continuous and discrete exponential signal**

### Explanation: Continuous and Discrete Exponential Signals

**Exponential signals** are fundamental in signal processing and system analysis. They are widely used to model natural phenomena such as growth and decay processes.

#### Continuous Exponential Signal

A continuous exponential signal is mathematically represented as:

$$
x(t) = A e^{\alpha t}
$$

where:
- \( A \) is the initial amplitude,
- \( \alpha \) is the exponential coefficient (rate of growth or decay),
- \( t \) is the continuous time variable.

- If \( \alpha > 0 \), the signal grows exponentially.
- If \( \alpha < 0 \), the signal decays exponentially.

#### Discrete Exponential Signal

A discrete exponential signal is given by:

$$
x[n] = A e^{\alpha n}
$$

where:
- \( n \) is the discrete time index.

The behavior is similar to the continuous case, but the signal is defined only at integer values of \( n \).

#### Applications

- Exponential signals are used to describe charging and discharging in capacitors, population growth, radioactive decay, and many other physical processes.
- In digital signal processing, discrete exponential signals are important for analyzing system stability and frequency response.

#### Visualization

- The **continuous exponential signal** appears as a smooth curve, either rising or falling depending on the sign of \( \alpha \).
- The **discrete exponential signal** is represented by distinct points (stems), showing the value at each sample.

These signals help in understanding the dynamic behavior of systems and are essential in both theoretical and practical aspects of engineering and science.

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_continuous_exponential(time, amplitude, coefficient):
    exponential_signal = amplitude * np.exp(coefficient * time)
    return exponential_signal

def simulate_discrete_exponential(num_samples, amplitude, coefficient):
    exponential_signal = amplitude * np.exp(coefficient * np.arange(num_samples))
    return exponential_signal

# Define the time range for the continuous exponential signal
time = np.linspace(0, 5, 1000)  # Time range from 0 to 5

# Define the number of samples, initial amplitude, and coefficient for the discrete exponential signal
num_samples = 20  # Number of samples
amplitude = 2  # Initial amplitude
coefficient = -0.5  # Exponential coefficient

# Simulate the continuous exponential signal
continuous_exponential = simulate_continuous_exponential(time, amplitude, coefficient)

# Simulate the discrete exponential signal
discrete_exponential = simulate_discrete_exponential(num_samples, amplitude, coefficient)

# Plot and display the continuous and discrete exponential signals
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(time, continuous_exponential)
plt.title('Continuous Exponential Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.stem(discrete_exponential)
plt.title('Discrete Exponential Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

# Save the exponential signal arrays (optional)
# np.savetxt('continuous_exponential.txt', continuous_exponential, delimiter=',')
# np.savetxt('discrete_exponential.txt', discrete_exponential, delimiter=',')
```

### **AIM: Write a python program to Simulate continuous and discrete parabolic signal**

### Explanation: Continuous and Discrete Parabolic Signals

**Parabolic signals** are important in signal processing and system analysis, representing signals with a quadratic (second-order polynomial) relationship.

#### Continuous Parabolic Signal

A continuous parabolic signal is mathematically represented as:

$$
x(t) = a t^2 + b t + c
$$

where:
- \( a, b, c \) are the coefficients of the quadratic equation,
- \( t \) is the continuous time variable.

- The signal forms a parabola, opening upwards if \( a > 0 \) and downwards if \( a < 0 \).
- The vertex and axis of symmetry depend on the coefficients.

#### Discrete Parabolic Signal

A discrete parabolic signal is given by:

$$
x[n] = a n^2 + b n + c
$$

where:
- \( n \) is the discrete time index.

- The signal consists of discrete points following a quadratic relationship.

#### Applications

- Parabolic signals are used to model acceleration in physics, such as the motion of objects under constant acceleration.
- They are also used in digital signal processing and control systems for testing and analysis.

#### Visualization

- The **continuous parabolic signal** appears as a smooth curve (parabola) over time.
- The **discrete parabolic signal** is shown as a sequence of points that follow the same quadratic trend.

Parabolic signals help in understanding non-linear system behaviors and are useful in both theoretical and practical engineering contexts.

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_continuous_parabolic(time, coefficients):
    parabolic_signal = np.polyval(coefficients, time)
    return parabolic_signal

def simulate_discrete_parabolic(num_samples, coefficients):
    parabolic_signal = np.polyval(coefficients, np.arange(num_samples))
    return parabolic_signal

# Define the time range for the continuous parabolic signal
time = np.linspace(-5, 5, 1000)  # Time range from -5 to 5

# Define the number of samples and coefficients for the discrete parabolic signal
num_samples = 20  # Number of samples
coefficients = [1, 2, 1]  # Coefficients of the parabolic signal

# Simulate the continuous parabolic signal
continuous_parabolic = simulate_continuous_parabolic(time, coefficients)

# Simulate the discrete parabolic signal
discrete_parabolic = simulate_discrete_parabolic(num_samples, coefficients)

# Plot and display the continuous and discrete parabolic signals
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(time, continuous_parabolic)
plt.title('Continuous Parabolic Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.stem(discrete_parabolic)
plt.title('Discrete Parabolic Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

# Save the parabolic signal arrays (optional)
# np.savetxt('continuous_parabolic.txt', continuous_parabolic, delimiter=',')
# np.savetxt('discrete_parabolic.txt', discrete_parabolic, delimiter=',')
```
