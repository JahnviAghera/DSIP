```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define threshold value and max value
thresh = 127
max_val = 255

# Define range of input values
x = np.arange(0, 256)

# Threshold logic for each type
y_binary = np.where(x > thresh, max_val, 0)
y_binary_inv = np.where(x > thresh, 0, max_val)
y_trunc = np.where(x > thresh, thresh, x)
y_tozero = np.where(x > thresh, x, 0)
y_tozero_inv = np.where(x > thresh, 0, x)

# Plotting all threshold types
plt.figure(figsize=(15, 8))

threshold_graphs = [
    ("BINARY", y_binary),
    ("BINARY_INV", y_binary_inv),
    ("TRUNC", y_trunc),
    ("TOZERO", y_tozero),
    ("TOZERO_INV", y_tozero_inv)
]

for i, (name, y) in enumerate(threshold_graphs):
    plt.subplot(2, 3, i + 1)
    plt.plot(x, y, color='blue')
    plt.title(name)
    plt.xlabel('Input Intensity')
    plt.ylabel('Output Intensity')
    plt.grid(True)
    plt.xlim(0, 255)
    plt.ylim(0, 255)

plt.suptitle("Input-Output Relationship for OpenCV Threshold Types")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
